```python
import sqlite3
from IPython.display import display
from aizynthfinder.reactiontree import ReactionTreeFromDict

def draw_tree_by_id(i: int, inverse=True):
    with sqlite3.connect('results.db') as conn:
        res = conn.execute(f'''
        SELECT tree_dict FROM synthetic_routes
        WHERE id = {i}
        ''').fetchall()[0][0]
    
    if res:
        tree_dict = eval(res)
        tree = ReactionTreeFromDict(tree_dict).tree
        display(tree.to_image(inverse=inverse))
        
draw_tree_by_id(1)
```

```python
solved_both_id =[3, 4, 7, 9, 10, 19, 24, 25, 30, 41, 46, 47, 59,
                 61, 62, 70, 87, 124, 167, 193, 317]
```

```python
for i in solved_both_id:  
    draw_tree_by_id(i)
```
