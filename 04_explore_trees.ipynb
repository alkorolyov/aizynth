```python
import sqlite3

import numpy as np
from IPython.display import display
from aizynthfinder.reactiontree import ReactionTreeFromDict

def draw_tree_by_id(i: int, inverse=True):
    with sqlite3.connect('results.db') as conn:
        res = conn.execute(f'''
        SELECT tree_dict FROM routes
        WHERE id = {i}
        ''').fetchall()[0][0]
    
    if res:
        tree_dict = eval(res)
        tree = ReactionTreeFromDict(tree_dict).tree
        display(tree.to_image(inverse=inverse))
        
def draw_tree(tree_dict, inverse=True):
    if isinstance(tree_dict, str):
        tree_dict = eval(tree_dict)    
    tree = ReactionTreeFromDict(tree_dict).tree
    display(tree.to_image(inverse=inverse))
```

```python
import gzip
import json
import pandas as pd

with gzip.open('results/output.json.gz', 'rb') as f:
    res = pd.DataFrame(json.load(f)['data'])

# with gzip.open('results/chembl_1000_az.json.gz', 'rb') as f:
#     res = json.load(f)
#     res = pd.DataFrame(res['data'])

solved = res[res.is_solved]
solved
```

```python
import numpy as np

idx = 3

cluster_labels = solved.cluster_labels.iloc[idx]

for cluster in np.unique(cluster_labels):
    first_occurrence = np.argwhere(cluster_labels == cluster)[0][0]
    tree = solved.iloc[idx]['trees'][first_occurrence]
    draw_tree(tree, inverse=True)
```
